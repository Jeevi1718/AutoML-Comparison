```python
from pycaret.regression import *
```

```python
import pandas as pd
import numpy as np
d=pd.read_csv('/content/drive/MyDrive/ML data set/diabetes.csv')
```

```python
d.isna().sum()
```

```python
d[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']]=d[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.nan)
```

```python
s=setup(d,target="Outcome")
```

```python
best=compare_models()
```

```python
print(best)
```

```python
evaluate_model(best)
```

```python
plot_model(best,plot="feature")
```

```python
predict_model(best)
```

```python
prediction=predict_model(best,data=d)
```
